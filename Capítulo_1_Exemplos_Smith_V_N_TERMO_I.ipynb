```python
import math as mt
m = 6.14 # [Kg]
g = 9.82 # [m/s²]
d = 1 # [cm]

F = m*g #[N]
pi = mt.pi
A = ((pi*(d**2))/4)

pmano = F/A
pbaro = 748 #[Torr]
patm = pbaro * 0.013332 # Fator de Conversão

pabs = pmano + patm
print(f"Pressão Absoluta: {(round(pabs, 2))} N/cm²")
```

```python
import math as mt
m = 2500 # [Kg]
z1 = 10  # [m]
z2 = 100 # [m]
g = 9.8 # [m/s²]

#a):

E = m*g*z1
print(f"Resultado da Energia Potencial: {(round(E))} J")

#b)

W = m*g*(z2-z1)
print(f"Trabalho Realizado: {(round(W))} J")

#c)

E2 = m*g*z2
print(f"Resultado da Energia Potencial: {(round(E2))} J")

#d)  dEK + dEP = 0 : Ekf - Eki + Epf - Epi = o ---> Ekf = Epi

Epi = E2
v = mt.sqrt((2*Epi/m))
print(f"Velocidade Final: {(round(v, 2))} m/s")


```
